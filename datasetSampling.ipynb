# Multithreading code starts here

```python
# SAMPLE FROM UNIQUE CHECKINS
# Maximum range is equal to the number of userIds
import threading 
lock = threading.Lock()
new_dataframe = pd.DataFrame()
```

```python
def add_checkins(userStartId,numberOfUsers):
    global new_dataframe
    global unique_checkins
    global lock
    for userid in range(userStartId, userStartId+numberOfUsers):
        #How many checkin do we want for this user?
        user_checkin_number = randrange(35,65)
        #Generating checkin for the user
        t0 = time.clock()
        for j in range(user_checkin_number):
            #Sampling the checkin from the checkins that we have originally
            sampled_checkin = unique_checkins[randrange(0,227164)]
            #Making a row to append later to dataframe
            new_row = {'User ID (anonymized)':userid,'Venue ID (Foursquare)':sampled_checkin[0],'Venue category ID (Foursquare)':sampled_checkin[1], 'Venue category name (Fousquare)':sampled_checkin[2],'Latitude':sampled_checkin[3], 'Longitude':sampled_checkin[4], 'UTC time':sampled_checkin[5],'Time':sampled_checkin[6]}
            lock.acquire()
            new_dataframe = new_dataframe.append(new_row, ignore_index=True)
            lock.release()
        t1 = time.clock()
        print(f"Data added for user number: {userid} and numbered: {user_checkin_number} and time {t1-t0} seconds.")
t0 = time.clock()

number_of_users = 1000
number_of_threads = 4
users_per_thread = int(number_of_users / number_of_threads)

threads = [threading.Thread(target=add_checkins, args=(users_per_thread*i+1,users_per_thread)) for i in range(number_of_threads)]
[thrd.start() for thrd in threads]
[thrd.join() for thrd in threads]
print(f"Time taken: {time.clock() - t0} seconds")
```
